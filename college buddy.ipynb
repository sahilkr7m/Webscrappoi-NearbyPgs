```python
import time
import timeit
from bs4 import BeautifulSoup
## need selenium to load the page for HTML to show
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

#initiate the driver
driver = webdriver.Chrome(ChromeDriverManager().install())

# Open the URL (link to your public google maps list)
url = 'https://goo.gl/maps/gUEbFVJm25t54BQE7'
driver.get(url)
```

```python
start = timeit.default_timer()
scrolling_element_xpath = '/html/body/div[3]/div[9]/div[8]/div/div[1]/div/div/div[5]'
scrolling_element= driver.find_element_by_xpath(scrolling_element_xpath)    

# use height of element to determine if need to scroll 
last_height = driver.execute_script("return arguments[0].scrollHeight", scrolling_element)
print(last_height)

SCROLL_PAUSE_TIME = 2.0 # pause before next scroll, to let page load 
t = 0 # number of times we have scrolled
while True:
    print(t)
    # Scroll down to bottom of whatever is currently loaded
    driver.execute_script('arguments[0].scrollTo(0, arguments[0].scrollHeight)', scrolling_element)
    t = t+1

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Check if more scrolling required 
    new_height = driver.execute_script("return arguments[0].scrollHeight", scrolling_element)
    print(new_height)
    if new_height == last_height:
        break
    last_height = new_height


stop = timeit.default_timer()

print('Time taken: ', stop - start) 
```

```python
def click_element(i):
    print('clicking item')
    ## Select the item using xpath 
    xpath = '//*[@id="pane"]/div/div[1]/div/div/div[5]/div[' +  str(i) + ']'
    element = driver.find_element_by_xpath(xpath)

    ## Click into listing
    element.click()
    time.sleep(5.0)
    
def back_to_list():
    print('going back to list')
    driver.execute_script("window.history.go(-1)")
    time.sleep(5.0)


## need this function to load listing to be visisble and selectable on sidebar. 
# where i is the ID of the element in the HTML
def load_listings(i):
    
    ## must scroll for every 20 (i = 40) listings
    x = (i-1)//40
    print('loading scrollbar...', x, ' times')
    
    # define scrolling bar
    scrolling_element_xpath = '/html/body/div[3]/div[9]/div[8]/div/div[1]/div/div/div[5]'
    scrolling_element = driver.find_element_by_xpath(scrolling_element_xpath)  

    for j in range(0, x):
        driver.execute_script('arguments[0].scrollTo(0, arguments[0].scrollHeight)', scrolling_element)
        time.sleep(2.0)
```

```python
from selenium.common.exceptions import NoSuchElementException

start = timeit.default_timer()

url = 'https://www.google.com/maps/@12.2914328,76.6528555,15z/data=!3m1!4b1!4m3!11m2!2suEgkHCF1SraV_3b6KUKlUg!3e3'
driver.get(url)
time.sleep(2.0)

urls = []
errors = 0 

for i in range(1, len(df)*2, 2): 
    
    print(i)
    
    try:
        if ((i-1)//40) > 0: 
            load_listings(i)
        click_element(i)
      
    except NoSuchElementException:  #if cannot find, wait another 5 sec
        errors += 1
        print("running NoSuchElementException...sleeping 5 sec")
        time.sleep(5.0)
        back_to_list()
        if ((i-1)//40) > 0: 
            load_listings(i)
        click_element(i)
        print("end of exception")
    
    # append URL
    element_url = driver.current_url
    urls.append(element_url)
    
    # go back to list and try next item
    back_to_list()



stop = timeit.default_timer()

print('Time taken: ', stop - start) 
print('Errors: ', errors)
```

```python
df['urls'] = urls

## use regex to extract the lat-lng from URL
extract_lat_lng = df['urls'].str.extract(r'!3d(-?\d+(?:\.\d+)?)!4d(-?\d+(?:\.\d+))')
## output will be 2 columns titled '0' and '1'.

df['lat'], df['lng'] =  extract_lat_lng[0], extract_lat_lng[1]

df.head()
# save list
df.to_csv('new_list.csv')
```
